```python
#1. Створіть одновимірний масив (вектор) із послідовністю чисел, де кожне наступне число більше попереднього на 3, починаючи з 5, та виведіть його значення.

import numpy as np
array = np.arange(5, 100, 3)
print(array)
```

```python
#2. Створіть двовимірний масив (матрицю) розміром 3х3, заповніть його добутками квадратів кожного числа та його подвійного значення, починаючи з 1, та виведіть його значення.

import numpy as np
rows, cols = 3, 3
matrix = np.zeros((rows, cols))

for i in range(rows):
    for j in range(cols):
        num = i * cols + j + 1  # Обчислюємо число, починаючи з 1
        square = num ** 2
        double = num * 2
        matrix[i, j] = square * double

print(matrix)
```

```python
#4. Створіть масив розміром 4х4, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть вектор-рядок, що містить середнє значення елементів кожного стовпця матриці.

import numpy as np
matrix = np.random.rand(4, 4)
column_means = matrix.mean(axis=0)
print("Матриця:\n", matrix)
print("Вектор-рядок:\n", column_means)
```

```python
#5. Створіть два одновимірних масиви розміром 7, заповніть їх випадковими дійсними числами від -2.5 до 2.5 та виконайте на них поелементні операції додавання, віднімання та множення.

import numpy as np
array1 = np.random.uniform(-2.5, 2.5, 7)
array2 = np.random.uniform(-2.5, 2.5, 7)

addition = array1 + array2
subtraction = array1 - array2
multiplication = array1 * array2

print("Перший масив:", array1)
print("Другий масив:", array2)
print("Додавання:", addition)
print("Віднімання:", subtraction)
print("Множення:", multiplication)
```

```python
#7. Створіть дві матриці розміром 4х2 та 2х2, заповніть їх випадковими цілими числами в діапазоні від 1 до 6 та перемножте їх між собою.

import numpy as np
matrix1 = np.random.randint(1, 7, (4, 2))
matrix2 = np.random.randint(1, 7, (2, 2))

product = np.dot(matrix1, matrix2)

print("Перша матриця:\n", matrix1)
print("Друга матриця:\n", matrix2)
print("Добуток матриць:\n", product)
```

```python
#8. Створіть матрицю розміром 2х2, заповніть її випадковими цілими числами в діапазоні від 1 до 12 та знайдіть її обернену матрицю.

import numpy as np
matrix = np.random.randint(1, 13, (2, 2))
print("Матриця:\n", matrix)

determinant = np.linalg.det(matrix)
if determinant == 0:
    print("Оберненої матриці не існує.")
else:
    inverse_matrix = np.linalg.inv(matrix)
    print("Обернена матриця:\n", inverse_matrix)
```

```python
#9. Створіть матрицю розміром 3х3, заповніть її випадковими цілими числами в діапазоні від 0 до 1 та транспортуйте її.

import numpy as np
matrix = np.random.randint(0, 2, (3, 3))
print("Матриця:\n", matrix)

transposed_matrix = np.transpose(matrix)
print("Транспортована матриця:\n", transposed_matrix)
```

```python
#13. Створіть дві матриці розміром 8х8, заповніть їх випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть різницю їх елементів.

import numpy as np
matrix1 = np.random.uniform(-1, 1, (8, 8))
matrix2 = np.random.uniform(-1, 1, (8, 8))

difference = matrix1 - matrix2

print("Перша матриця:\n", matrix1)
print("Друга матриця:\n", matrix2)
print("Різниця елементів матриць:\n", difference)
```

```python
#14. Створіть матрицю розміром 5х5, заповніть її випадковими дійсними числами і діапазоні від 0 до 1 та знайдіть середнє арифметичне значення елементів матриці.

import numpy as np
matrix = np.random.uniform(0, 1, (5, 5))

mean_value = np.mean(matrix)

print("Матриця:\n", matrix)
print("Середнє арифметичне значення елементів матриці:", mean_value)
```

```python
#15. Створіть матрицю розміром 9х9, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить найбільше за модулем значення елементу у кожному рядку матриці.

import numpy as np
matrix = np.random.uniform(-0.5, 0.5, (9, 9))
max_abs_values = np.max(np.abs(matrix), axis=1)
max_abs_values_column = max_abs_values.reshape(-1, 1)

print("Матриця:\n", matrix)
print("Вектор-стовпчик, що містить найбільше за модулем значення елементу у кожному рядку матриці:\n", max_abs_values_column)
```
